```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
#read data set from csv
data_set = pd.read_csv('Salary_Data.csv')
data_set
```

```python
data_set.info()
```

```python
data_set.describe()
```

```python
data_set.columns.isna().sum()
```

```python
X = data_set.iloc[:,0:1]
Y = data_set.iloc[:,1:]
X
```

```python
Y
```

```python
#Splitting datset into Training set and Testing set
from sklearn import model_selection
```

```python
#Splitting datset into Training set and Testing set
X_train , X_test,Y_train ,Y_test = model_selection.train_test_split(X,Y,test_size=0.3,random_state=101)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
#fitting Simple linear regression to the train data set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor
```

```python
#fit the train dataset
model = regressor.fit(X_train , Y_train)
model
```

```python
#Predicting the test data set
pred_ds = model.predict(X_test)
```

```python
pd.DataFrame(pred_ds)
```

```python
Y_test
```

```python
#visualize the Training set results
import matplotlib.pyplot as plt
```

```python
#visualize the Training set results
plt.scatter(X_train , Y_train,color='red')
plt.plot(X_train , model.predict(X_train),color='blue')
plt.title = 'Salary vs Experience(Training set )'
plt.xlabel('Experience')
plt.ylabel = ('Salary')
plt.show()
```

```python
#visualize the Test set results
plt.scatter(X_test , Y_test,color='red')
plt.plot(X_train , model.predict(X_train),color='blue')
plt.plot(X_test, pred_ds,color="black")
plt.title = 'Salary vs Experience(Training set )'
plt.xlabel('Experience')
plt.ylabel = ('Salary')
plt.show()
```
